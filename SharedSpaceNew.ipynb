```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import cudf
import cuml
import cupy as cp
```

```python
cudf_df_male = cudf.DataFrame.from_pandas(male_df)
cudf_df_female = cudf.DataFrame.from_pandas(female_df)
```

KMeans

```python
import time
from sklearn.cluster import KMeans
start_time_cpu = time.time()
kmeans_cpu = KMeans(n_clusters=30)
labels_male_kmeans_cpu = kmeans_cpu.fit_predict(male_df)
end_time_cpu = time.time()
runtime_cpu = end_time_cpu - start_time_cpu
```

```python
labels_male_kmeans_cpu
```

```python
from cuml.cluster import KMeans
start_time_gpu = time.time()
kmeans = KMeans(n_clusters=30)
cudf_df_male_float32 = cudf_df_male.astype(np.float32)
labels_male_kmeans_rapids = kmeans.fit_predict(cudf_df_male_float32)
cp.cuda.Stream.null.synchronize()
end_time_gpu = time.time()
runtime_gpu = end_time_gpu - start_time_gpu
```

```python
labels_male_kmeans_rapids
```

```python
print("CPU Runtime:", runtime_cpu)
print("GPU Runtime:", runtime_gpu)
speedup = runtime_cpu / runtime_gpu
print("Speedup:", speedup)
```

Agglomerative Clustering

```python
cupy_array_male = cp.array(data_male)
```

```python
from cuml.cluster import AgglomerativeClustering
start_time_gpu_agglo = time.time()
agg_clustering = AgglomerativeClustering(n_clusters=30)
labels_male_agglo_rapids = agg_clustering.fit_predict(cupy_array_male)
cp.cuda.Stream.null.synchronize()
end_time_gpu_agglo = time.time()
runtime_gpu_agglo = end_time_gpu_agglo - start_time_gpu_agglo
```

```python
from sklearn.cluster import AgglomerativeClustering
start_time_cpu_agglo = time.time()
agglomerative_cpu = AgglomerativeClustering(n_clusters=30)
labels_male_agglo_cpu = agglomerative_cpu.fit_predict(data_male)
end_time_cpu_agglo = time.time()
runtime_cpu_agglo = end_time_cpu_agglo - start_time_cpu_agglo
```

```python
labels_male_agglo_cpu
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
linkage_matrix = linkage(data_male, method='ward')
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Index')
plt.ylabel('Distance')
plt.show()
```

```python
print("CPU Runtime:", runtime_cpu_agglo)
print("GPU Runtime:", runtime_gpu_agglo)
speedup_agglo = runtime_cpu_agglo / runtime_gpu_agglo
print("Speedup:", speedup_agglo)
```
